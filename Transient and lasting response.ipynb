```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal
```

***Q1***

```python
a = [1,0.5]
b = [2]
x = np.concatenate([np.zeros(5) ,np.ones(15)])
y = signal.lfilter(b,a,x)
plt.figure(figsize=(15,4.8))
plt.plot(x, 'r-o')
plt.plot(y, 'b-o')
```

```python
a = [1,-0.7,0.1]
b = [2, 0, -1]
x = np.concatenate([np.zeros(5) ,np.ones(15)])
y = signal.lfilter(b,a,x)
plt.figure(figsize=(15,4.8))
plt.plot(x, 'r-o')
plt.plot(y, 'b-o')
```
